## Sorting majors by Popularity
I based the popularity of each major category by the number or count of graduates 

```python
majors = recent_grads["Major_category"].value_counts()
majors.describe()
```

```python
#Based from the mean
popular_majors = majors[majors > 10].index
unpopular_majors = majors[majors<10].index

popular_major_df = recent_grads[recent_grads["Major_category"].isin(popular_majors)]
popular_major_median = popular_major_df["Median"]
popular_major_median.head()

plt.pie(majors.head())
plt.legend(labels = list(popular_major_df["Major"].head()),bbox_to_anchor = [1,0.5])
```

```python
unpopular_major_df = recent_grads[recent_grads["Major_category"].isin(unpopular_majors)]
unpopular_major_median = unpopular_major_df["Median"]
unpopular_major_median.head()

plt.pie(majors.tail())
plt.legend(labels = list(unpopular_major_df["Major"].head()),bbox_to_anchor = [1,0.5])
```

```python
recent_grads.sort_values("Unemployment_rate").tail(10).plot.barh(x = 'Major',y = 'Unemployment_rate', legend = False)
recent_grads.sort_values("Unemployment_rate").head(10).plot.barh(x = 'Major', y= 'Unemployment_rate', legend= False)
```

```python
recent_grads.sort_values("ShareWomen").tail(10).plot.barh(x = 'Major',y = 'ShareWomen', legend = False)
plt.title("Major with most women")

recent_grads.sort_values("ShareWomen", ascending = False).tail(10).plot.barh(x = 'Major', y= 'ShareWomen', legend= False)
plt.title("Major with least women")
```

```python
fig,ax = plt.subplots()
ax.hist(unpopular_major_median)
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax1.hist(popular_major_median,bins=30,range=(0,100000))
ax1.set_xlabel("Median")
ax1.set_ylabel("Count")
ax1.set_title("Distribution of Median for Popular Majors")
ax2 = fig.add_subplot(1,2,2)
ax2.hist(unpopular_major_median,bins=30,range=(0,100000))
ax2.set_xlabel("Median")
ax2.set_ylabel("count")
ax2.set_title("Distibution of Median for Unpopular Majors")
```

Do students in more popular major make more money?

    2. The highest median for unpopular major does not exceed in 60 000.
    3. Though, the most count of popular major and unpopular major is in between 20000 - 40000, popular major has still
        greater numbers in terms of greater median since the population of unpopular major is less.
        

```python
median_male_majority = recent_grads[recent_grads["Women"] < recent_grads["Men"]]["Median"]
median_female_majority = recent_grads[recent_grads["Women"] > recent_grads["Men"]]["Median"]
```

```python
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax1.hist(median_male_majority,bins=30,range=(0,100000))
ax1.set_xlabel("Median")
ax1.set_ylabel("Frequency")
ax1.set_title("Distribution of Median Salary, Number of Women < Men Major")
ax2 = fig.add_subplot(1,2,2)
ax2.hist(median_female_majority,bins=30,range=(0,100000))
ax2.set_xlabel("Median")
ax2.set_ylabel("Frequency")
ax2.set_title("Distribution of Median Salary, Number of Women > Men Major")
```

1. The highest median where number of men is greater is greater than 70,000 so by means, Men makes the highest median
2. Disregard the frequency, the high medians where number of men is greater are scattered between 30,000 to 50,000
3. Disregard the frequency, the high medians where number of women is greater are compressed in 30,000 more or less.
